## DATA CLEANING

#### we will now check for null values using the .isnull() fucntion,.sum() is used to sum up all the nulls in individual columns

```python
thyroid.isnull().sum()
```

#### this returns that there are zero nulls in the dataset but we know this to be false since we discovered above that they are not labeled as NANs or 0 but rather as '?' 
#### we will need to replace them with NAN's 

#### first let us replace all the '?' values with NAN's  using the replace method and np.nan

```python
thyroid = thyroid.replace(['?'],np.nan)
```

#### we will run .isnull a second time and check for nulls

```python
thyroid.isnull().sum()
```

#### above we can see that the dataset has a high amount of null values after replacing the '?' values with NAN's
#### the last column 'TBG' is made up of mostly null values thus conducting median imputation would skew that column's data too much for it to be used to create a model
#### we will not drop these values and instead proceed to do modelling with them 
#### we will train and test the models with the NAN's and later do imputation with the median values of the columns and compare the performance of the different models

#### NB:we tried to train the model with the NAN's but got an error instead,we will instead conduct imputation on the dataset using the median for the numeric columns and the mode for categorical columns
#### NB2: we changed the median imputation of the age column to mean imputation,explanation is given below

#### proceeding on to check for duplicated rows in the dataset using the duplicated() method

```python
print(thyroid.duplicated().sum())  
```

#### as we observed above after running pandas_profiling the dataset contains duplicated rows,we will proceed to drop these using the drop_duplicates() method and then check to see if all were dropped

```python
thyroid = thyroid.drop_duplicates()
```

```python
thyroid.duplicated().sum()
```

#### we will now conduct imputation using sklearn's impute class to replace all the numerical column NaN's with their median values

```python
numerics = ['FTI','T3','T4U','TBG','TSH','TT4']

imputer = SimpleImputer(strategy = 'median')

thyroid[numerics] = imputer.fit_transform(thyroid[numerics])
```

#### now to replace the categorical and boolean columns with the mode values

```python
strings = ['status','sex','on_thyroxine','query_on_thyroxine','FTI_measured','lithium','on_antithyroid_medication',
            'pregnant','query_hypothyroid','query_hyperthyroid','sick','T3_measured','T4U_measured','TBG_measured',
            'thyroid_surgery','TSH_measured','TT4_measured','tumor']

imputer2 = SimpleImputer(strategy='most_frequent')

thyroid[strings] = imputer2.fit_transform(thyroid[strings])
```

#### we will now check whether we were succesfull by running the isnull function to count the total number of nulls
#### if the method returns all zeros for all the columns then we were succesfull

```python
thyroid.isnull().sum()
```

#### as you can see above all the columns display a total null count of zero indicating that we were succesful in eliminating all the null values
#### NB the null values in the age column are due to the fact that we did conduct median imputation on that  column and instead opted for mean imputation

```python
thyroid.dtypes
```

#### we will also need to change the datatypes of the columns as we observed all of them are objects,and conducting EDA with objects will prove difficult

#### we will create a for loop to change all the column dtypes to either categorical,numerical or boolean dtypes and print out the datatypes

```python
categorical_cols = ['on_thyroxine','query_on_thyroxine','FTI_measured','lithium','on_antithyroid_medication',
            'pregnant','query_hypothyroid','query_hyperthyroid','sick','T3_measured','T4U_measured','TBG_measured',
            'thyroid_surgery','TSH_measured','goitre','TT4_measured','tumor','status','sex']

numerical_cols = ['age','FTI','T3','T4U','TBG','TSH','TT4']

for col in thyroid.columns:
    if col in numerical_cols:
        thyroid[col] = pd.to_numeric(thyroid[col])
    
    elif col in categorical_cols:
        thyroid[col] = thyroid[col].astype('category')
    

thyroid.dtypes
```

```python
thyroid['age']= thyroid['age'].fillna(thyroid['age'].mean(), axis = 0)
```

#### above we imputed the nulls in the age column with the mean of that column

## MODELLING

### Random Forest

#### to perform modelling we will perform minimal feature engineering which will include running the get dummies command on the training 
#### this will perform label encoding on the features 

```python
thyroid.head()
```

#### we will drop the dependent variable to create our independent features x

```python
x = thyroid.drop(['status'],axis = 1)
```

#### using get_dummies to perform label encoding

```python
x = pd.get_dummies(x)
```

```python
x
```

#### we can observe that the categorical columns have been converted to integer format,which will make training the dataset easier

```python
y = thyroid['status']
```

```python
y = y.cat.codes
```

#### above we used the cat.codes method to convert the status column values to categorical values like we did for the independent features

```python
y
```

#### below we will split the data into training and testing sets, the test size size will be 30% of the data

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
forest = RandomForestClassifier(n_estimators = 200, random_state=42, min_samples_split=10, max_depth=15)

forest = forest.fit(xtrain, ytrain)

ypred = forest.predict(xtest)

f1_score(ytest, ypred)
```

#### above we created a baseline random forest classifier model with 200 estimators and a max depth of 15 
#### this yielded us an accuracy of 99.6% which is above our metric for success of 85%

```python
tree = forest.estimators_[20]
importances = list(tree.feature_importances_)

feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(x.columns, importances)]                   
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

                   
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
sns.set(style = 'whitegrid')

plt.figure(dpi = 100, figsize = (15,5))

x_values = list(range(len(importances)))

plt.bar(x_values, importances, orientation = 'vertical')

plt.xticks(x_values, x.columns, rotation= 'vertical')


plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
```

#### the feature importances suggest that the TT4 column is the most important variable in predicting whether a specific individual will have hypothyroidism

```python
dot_data = StringIO()
tree = forest.estimators_[20]

export_graphviz(tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = x.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Hypothyroid_Test.png')
Image(graph.create_png())
```

## Gradient Boosting

## Ada Boost Classsifier

```python
adc = AdaBoostClassifier(DecisionTreeClassifier(max_depth=5),n_estimators=250)
                         

adc = adc.fit(xtrain, ytrain)

y_pred = adc.predict(xtest)

f1_score(ytest, y_pred)
```

## Support Vector Machine

```python
svc = SVC(kernel = 'linear',  C = 1, gamma = 0.001)

svc = svc.fit(xtrain, ytrain)


y__pred = svc.predict(xtest)

f1_score(ytest, y__pred)
```

```python
svc = SVC(kernel = 'poly', C = 1, gamma = 0.001)

svc.fit(xtrain, ytrain)


Y_pred = svc.predict(xtest)

f1_score(ytest, Y_pred)
```

```python
svc = SVC(kernel = 'rbf', C = 4, gamma = 0.001)

svc.fit(xtrain, ytrain)


y_pred = svc.predict(xtest)

f1_score(ytest, y_pred)
```

### perfroming hyperparameter tuning support vector machine

```python
params = {'C': [0.001, 0.01, 0.1, 1, 10],
          'gamma': [0.001, 0.01, 0.1, 1]}

svc = SVC()

search = RandomizedSearchCV(estimator = svc, 
                      param_distributions = params, 
                      scoring = 'accuracy',
                      n_jobs = 1, 
                      cv = 10 )
                
                
search.fit(x,y)
search.best_params_
```
